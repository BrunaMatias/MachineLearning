# Redução de dimensionalidade de imagens



```python
# Importação de bibliotecas para manipulação da imagem
import imageio
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem colorida
imagem_colorida = imageio.imread('/content/ec20deed774ea878d1d8ea95fe31e53adbe180e1a5f9d0e1449179a69cd33630._SX1080_FMjpg_.jpg')

# Imprime imagem colorida
plt.imshow(imagem_colorida)
plt.show()
```

```python
# Verifica se a imagem é RGB
if len(imagem_colorida.shape) == 3 and imagem_colorida.shape[2] == 3:
    # Converte para níveis de cinza usando fórmula da luminosidade dos 255 tons
    imagem_cinza = np.dot(imagem_colorida[..., :3], [0.299, 0.587, 0.114])
else:
    raise ValueError('A imagem não parece ser uma imagem RGB.')

# Converte os valores para inteiros de 8 bits (0-255)
imagem_cinza = imagem_cinza.astype(np.uint8)

# Exibe a imagem em níveis de cinza
plt.imshow(imagem_cinza, cmap='gray')
plt.show()
```

```python
# Define um limiar para converter para preto e branco
limiar = 128
imagem_preto_branco = np.where(imagem_cinza >= limiar, 255, 0).astype(np.uint8)

# Exibe imagem em preto e branco
plt.imshow(imagem_preto_branco, cmap='gray')
plt.show()
```
